```python
import os

import numpy as np
import matplotlib.pyplot as plt

import data_io as io
import beats
```

```python
path = os.path.join(os.path.curdir, 'Audio Analysis')
fname = 'Manic Polka.full.yml'

f = os.path.join(path, fname)
d, m = io.read(f)

segments = d['segments']
print(len(segments))
```

```python
segments[0]
```

```python
# Display
fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(1,1,1)

x = []
y = []
for s in segments:
    t0 = s['start']
    t1 = s['start'] + s['loudness_max_time']

    p0 = s['loudness_start']
    p1 = s['loudness_max']
    x.append(t0)
    x.append(t1)
    y.append(p0)
    y.append(p1)
    
ax.plot(x, y)
ax.set_xlim(0, 5)
plt.tight_layout()
plt.show()
```
