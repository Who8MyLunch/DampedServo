```python

import os
import numpy as np

import matplotlib.pyplot as plt

import beats
```

```python
path_data = os.path.curdir

fname_song = 'Rocket'

time_start = 22.
time_extract = 10.
```

```python
########################################
# Do it.
f = os.path.join(path_data, fname_song + '.wav')
data, sample_rate = beats.read_wav(f)

# Mono.
dt = data.dtype
data = data.mean(axis=1).astype(np.float32)
data = np.round(data).astype(dt)

# Extract data subset.
time_end = time_start + time_extract

k_start = int(sample_rate * time_start)
k_end = int(sample_rate * time_end)

data = data[k_start:k_end]

# Time samples.
dt = 1. / sample_rate
t = np.arange(time_start, time_end, dt)
t = t[:len(data)]

f = os.path.join(path_data, fname_song)
data_beats, data_segments = beats.analyze_song(f)
```

```python
########################################
# Display.
lo, hi = data.min(), data.max()
hi *= 1.05

fig = plt.figure(figsize=(24,6))
fig.clear()

ax = fig.add_subplot(1, 1, 1)
ax.plot(t, data, color='blue', linewidth=0.5)
    
for x in data_beats:
    if x >= time_end:
        break
    ax.vlines(x, lo, hi, color='orange', linewidth=2)
    
x = [v[0] for v in data_segments]    
y = [v[2] for v in data_segments]    
    
x = np.asarray(x)
y = np.asarray(y)
y -= y.min()
y /= y.max()

y *= (hi-lo)
y += lo

ax.plot(x, y, color='red', linewidth=3)
ax.plot(x, y, 'o', color='orange')
    
ax.set_xlim(t.min(), t.max())
ax.set_ylim(lo, hi)    
    
ax.set_xlabel('Time (sec)')
    
```
